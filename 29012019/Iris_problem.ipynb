```python
testing = pd.read_csv('./iris_test.csv')
```

```python
testing
```

```python
testing.loc[testing['species'] == 'Iris-setosa' ,'species'] = 0
testing.loc[testing['species'] == 'Iris-versicolor' ,'species'] = 1
testing.loc[testing['species'] == 'Iris-virginica' ,'species'] = 2
testing = testing.apply(pd.to_numeric)
```

```python
testing = testing.as_matrix()
```

```python
testing
```

```python
X_test = testing[:,:4]
y_test = testing[:,4]
```

```python
def build_model():
    net = keras.Sequential()
    net.add(keras.layers.Dense(10,input_shape=[4]))
    net.add(keras.layers.Activation('relu'))
    net.add(keras.layers.Dense(3))
    net.add(keras.layers.Activation('softmax'))
    sgd = keras.optimizers.SGD(lr=0.0005,decay=0,nesterov=True)
    net.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])
    return net     
```

```python
model = build_model()
y_train =keras.utils.to_categorical(y_train,3)
y_test = keras.utils.to_categorical(y_test,3)
```

```python
y_train
```

```python
model.fit(X_train,y_train,batch_size=10,epochs=7000,validation_data=(X_test,y_test),shuffle=True)
```

```python
y_test
```

```python
y_pre = model.predict_classes(X_test)
```

```python
y_pre = keras.utils.to_categorical(y_pre,3)
```
